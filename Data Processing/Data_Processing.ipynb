```python
df['World Rank'] = df['World Rank'].map(lambda x: x.lstrip('='))
df.tail(10)
```

```python
new_cols = ['World Rank',
 'University Name',
 'City',
 'Country',
 'Overall Score',
 'Academic Reputation',
 'Employer Reputation',
 'Citations per Faculty',
 'Faculty Students Ratio',
 'International Students Ratio',
 'International Faculty Ratio',
 'International Research Network',
 'Employment Outcomes'
 ]
```

```python
df['City'] = df['Country'].str.split(',').apply(lambda x: x[0])
df['Country'] = df['Country'].str.split(',').apply(lambda x: x[-1])
df=df.reindex(columns= new_cols)
df
```

```python
columns = ['Overall Score', 'Academic Reputation','Employer Reputation', 'Citations per Faculty', 'Faculty Students Ratio', 'International Students Ratio', 'International Faculty Ratio', 'International Research Network', 'Employment Outcomes']
df[columns] =df[columns].replace('-', '0')
df[columns] =df[columns].replace('QS Stars', NaN)
# df[columns] =df[columns].replace('QS Stars', 0)
df[columns]=df[columns].apply(pd.to_numeric)
df['Country'] = df['Country'].replace('5', '')
df.info()
```

```python
df1 = pd.read_csv("D:\Work Station\Scraper\Data_new1.csv")
df2 = pd.read_csv("D:\Work Station\Scraper\Data_new2.csv")
df3 = pd.read_csv("D:\Work Station\Scraper\Data_new3.csv")
df4 = pd.read_csv("D:\Work Station\Scraper\Data_new4.csv")
df5 = pd.read_csv("D:\Work Station\Scraper\Data_new5.csv")
df6 = pd.read_csv("D:\Work Station\Scraper\Data_new6.csv")
df7 = pd.read_csv("D:\Work Station\Scraper\Data_new7.csv")
```

```python
frames = [df1, df2, df3, df4, df5, df6, df7]
result1 = pd.concat(frames, axis=0)
result1
```

```python
df = df.reset_index(drop=True)
result1 = result1.reset_index(drop=True)
result2 = pd.concat([df, result1], axis=1)
result2.drop('Unnamed: 0', inplace=True, axis= 1)
result2.head()
# result_final.shape
```

```python
result2.index = np.arange(1, len(result2)+1)
```

```python
col2 = ['UG student % (Domestic)','PG student % (Domestic)','UG student % (International)','PG student % (International)','Domestic staff %', 'International staff %']
result2[col2] = result2[col2].replace('%','')
# result2[col2] = result2[col2].replace('%','0')
result2['UG student % (Domestic)'] = result2['UG student % (Domestic)'].map(lambda x: x.rstrip('%'))
result2['PG student % (Domestic)'] = result2['PG student % (Domestic)'].map(lambda x: x.rstrip('%'))
result2['UG student % (International)'] = result2['UG student % (International)'].map(lambda x: x.rstrip('%'))
result2['PG student % (International)'] = result2['PG student % (International)'].map(lambda x: x.rstrip('%'))
result2['Domestic staff %'] = result2['Domestic staff %'].map(lambda x: x.rstrip('%'))
result2['International staff %'] = result2['International staff %'].map(lambda x: x.rstrip('%'))
```

```python
result2.info()
```

```python
col3 = ['Total students count','UG student % (Domestic)','PG student % (Domestic)','International students count','UG student % (International)','PG student % (International)','Domestic staff %', 'International staff %','Total faculty staff count']

result2[col3] = result2[col3].astype(str)
result2[col3] = result2[col3].apply(lambda x: x.str.replace(',', ''))

result2[col3] = result2[col3].apply(pd.to_numeric, errors='coerce')
```
